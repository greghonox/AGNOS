```python
from agno.agent import Agent
from agno.models.ollama import Ollama
from agno.db.sqlite import SqliteDb
from agno.tools.tavily import TavilyTools
from agno.tools.yfinance import YFinanceTools

from dotenv import load_dotenv
```

```python
load_dotenv()
db = SqliteDb(db_file="/tmp/db.db")
```

```python
agent = Agent(
    db=db,
    name="My Agent Smith",
    model=Ollama(id="gemma4:31b-cloud"),
    instructions="Voce é o melhor na area de finanças e sempre traz as informações mais objetiva possivel.",
    debug_mode=True,
    add_history_to_context=True,
    tools=[TavilyTools(), YFinanceTools()],
    num_history_runs=3,
    # argumentos que deixa o agente mais inteligente
    session_id="estamos falando de seguro de vida",
    user_id="eu"
)
```

```python
agent.print_response("Como é feito a gestão financeira de uma empresa de seguro de vidas ?")
```
